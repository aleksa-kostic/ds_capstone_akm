```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, random_split, TensorDataset
import numpy as np
import matplotlib.pyplot as plt


from sklearn.preprocessing import MinMaxScaler
import pandas as pd

# Assuming X_bal is your DataFrame
scaler = MinMaxScaler()

# Fit the scaler on your data and transform it
X_bal_normalized = pd.DataFrame(scaler.fit_transform(X_bal), columns=X_bal.columns)


# Split data into training, testing, and validation sets
total_samples = len(X_bal_normalized)
train_size = int(0.8 * total_samples)
val_size = (total_samples - train_size) // 2
test_size = total_samples - train_size - val_size

#print(y_bal.shape, y.shape)
X_tensor = torch.tensor(X_bal_normalized.values, dtype=torch.float32)
y_tensor = torch.tensor(y_bal.values, dtype=torch.float32)

dataset = TensorDataset(X_tensor, y_tensor)
train_data, val_data, test_data = random_split(dataset, [train_size, val_size, test_size])

train_loader = DataLoader(train_data, batch_size=32, shuffle=True)
val_loader = DataLoader(val_data, batch_size=32)
test_loader = DataLoader(test_data, batch_size=32)

# Define the RBF neural network
input_dim = X_bal_normalized.shape[1]
hidden_dim = 16  # Adjust as needed
output_dim = 1   # Assuming binary classification

class RBFNN(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(RBFNN, self).__init__()
        self.hidden_dim = hidden_dim
        self.centers = nn.Parameter(torch.rand(hidden_dim, input_dim))
        self.beta = nn.Parameter(torch.rand(hidden_dim))
        self.linear = nn.Linear(hidden_dim, hidden_dim)
        self.linear2 = nn.Linear(hidden_dim, hidden_dim)
        self.drop = nn.Dropout(p=0.2)
        self.linear3 = nn.Linear(hidden_dim, output_dim)

    def radial_basis_function(self, x):
        return torch.exp(-self.beta * (x.unsqueeze(1) - self.centers).pow(2).sum(2))

    def forward(self, x):
        rbf_out = self.radial_basis_function(x)
        h1 = self.linear(rbf_out)
        h2 = self.linear2(self.drop(h1))
        output = self.linear3(h2)
        return output, rbf_out

# Use GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model = RBFNN(input_dim, hidden_dim, output_dim).to(device)

# Loss function and optimizer
criterion = nn.BCEWithLogitsLoss()
optimizer = optim.Adam(model.parameters(), lr=0.005)

# Training loop
num_epochs = 50
train_losses = []
val_losses = []
train_accuracies = []
val_accuracies = []

for epoch in range(1, num_epochs + 1):
    print("Epoch", epoch)
    model.train()
    total_correct_train = 0
    total_samples_train = 0
    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs, _ = model(inputs)
        loss = criterion(outputs.squeeze(), labels)
        
        # Check for nan loss
        if torch.isnan(loss):
            raise ValueError("Training halted. NaN loss detected.")

        loss.backward()
        
        # Check for exploding gradients
        if any(torch.isnan(param.grad).any() for param in model.parameters()):
            raise ValueError("Training halted. Exploding gradients detected.")
        
        optimizer.step()

        # Calculate training accuracy
        predictions = torch.sigmoid(outputs).round().squeeze()
        total_correct_train += (predictions == labels).sum().item()
        total_samples_train += len(labels)

    train_accuracy = total_correct_train / total_samples_train
    train_accuracies.append(train_accuracy)
    train_losses.append(loss.item())

    # Validation
    model.eval()
    total_correct_val = 0
    total_samples_val = 0
    val_loss = 0
    with torch.no_grad():
        for inputs, labels in val_loader:
            inputs, labels = inputs.to(device), labels.to(device)

            outputs, _ = model(inputs)
            loss = criterion(outputs.squeeze(), labels)
            val_loss += loss.item()

            # Calculate validation accuracy
            predictions = torch.sigmoid(outputs).round().squeeze()
            total_correct_val += (predictions == labels).sum().item()
            total_samples_val += len(labels)

    val_accuracy = total_correct_val / total_samples_val
    val_accuracies.append(val_accuracy)
    val_losses.append(val_loss / len(val_loader))

    # Save model state_dict every 2 epochs
    if epoch % 2 == 0:
        torch.save(model.state_dict(), f'./rbf_nn_models/rbf_nn_epoch_{epoch}')

    print(f'Epoch {epoch}/{num_epochs}')
    print(f'Training Loss: {loss.item():.4f}, Training Accuracy: {train_accuracy:.4f}')
    print(f'Validation Loss: {val_loss / len(val_loader):.4f}, Validation Accuracy: {val_accuracy:.4f}')
    print()

# Plot losses and accuracies over epochs
plt.figure(figsize=(10, 5))

# Losses
plt.subplot(1, 2, 1)
plt.plot(range(1, num_epochs + 1), train_losses, label='Train')
plt.plot(range(1, num_epochs + 1), val_losses, label='Validation')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss Over Epochs')
plt.legend()
plt.grid(True)

# Accuracies
plt.subplot(1, 2, 2)
plt.plot(range(1, num_epochs + 1), train_accuracies, label='Train')
plt.plot(range(1, num_epochs + 1), val_accuracies, label='Validation')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Accuracy Over Epochs')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

# Testing

```python
TP = 0
TN = 0
FP = 0
FN = 0


with torch.no_grad():

    total_correct_test = 0
    total_samples_test = 0
    for inputs, labels in test_loader:
        inputs, labels = inputs.to(device), labels.to(device)

        outputs, _ = model(inputs)        

        # Calculate validation accuracy
        predictions = torch.sigmoid(outputs).round().squeeze()
        #print("PR3D", predictions.shape, type(predictions))
        #print("LABEL", labels.shape, type(labels))
        total_correct_test += (predictions == labels).sum().item()
        total_samples_test += len(labels)

        for i in range(len(labels)):
            if labels[i] == 1 and predictions[i] == 1:
                TP += 1
            elif labels[i] == 1 and predictions[i] == 0:
                FN += 1
            elif labels[i] == 0 and predictions[i] == 1:
                FP += 1
            elif labels[i] == 0 and predictions[i] == 0:
                TN += 1
            else:
                raise ValueError("Idiot.")

                
precision = TP / (TP + FP) 
recall = TP / (TP + FN)
f1 = (2 * precision * recall) / (precision + recall)
accuracy =  (TP + TN)/(TP + FP + TN + FN)

print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")
print(f"Accuracy: {accuracy}")
```
