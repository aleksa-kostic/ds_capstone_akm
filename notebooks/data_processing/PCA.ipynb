```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('cleaned_augmented_2.csv', index_col = 0)
df
```

### PCA

```python
pca = PCA()
pca.fit(df)
cumsum = np.cumsum(pca.explained_variance_ratio_)
```

```python
d = np.argmax(cumsum >= 0.95) + 1
d
```

```python
pca = PCA(n_components=d)
pca.fit(df)
```

```python
# Get the indices of the top features for each component
top_feature_indices = [component.argsort()[-1::-1][:1] for component in pca.components_]

# Get the feature names for the top features of each component
feature_names = list(df.columns[:-1])  
top_features_names = []

for indices in top_feature_indices:
    top_feature_names = [feature_names[i] for i in indices]
    top_features_names.append(top_feature_names)
```

```python
top_features_names
```

```python
pca = PCA(n_components=d)
pca.fit(df)
```

```python
x_pca = pca.transform(df)
```

```python
transformed_df = pd.DataFrame(x_pca, columns=['PC{}'.format(i+1) for i in range(d)])

transformed_df.columns = ['{}_{}'.format(top_features_names[i][0], transformed_df.columns[i]) for i in range(len(top_features_names))]
```

```python
transformed_df
```

```python
transformed_df.to_csv('pca_transformed_data.csv')
```
