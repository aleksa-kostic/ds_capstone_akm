```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cleaned_augmented_2.csv')
```

```python
df
```

```python
df.info()
```

```python
df.describe()
```

```python
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns[:25]  

plt.figure(figsize=(15, 8))
sns.boxplot(data=df[numerical_columns], orient="h")
plt.title('Box Plot of First Part of the Data')
plt.xlabel('Values')
plt.ylabel('Features')
plt.show()
```

```python
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns[25:]  

plt.figure(figsize=(15, 8))
sns.boxplot(data=df[numerical_columns], orient="h")
plt.title('Box Plot of Second Part of the Data')
plt.xlabel('Values')
plt.ylabel('Features')
plt.show()
```

```python
numerical_columns_subset = df.select_dtypes(include=['int64', 'float64']).columns[:15]  

# Create histograms for the subset of numerical columns
df[numerical_columns_subset].hist(bins=20, figsize=(15, 10))
plt.suptitle('Histograms of the Columns -Part 1')
plt.show()
```

```python
numerical_columns_subset = df.select_dtypes(include=['int64', 'float64']).columns[15:31]  

# Create histograms for the subset of numerical columns
df[numerical_columns_subset].hist(bins=20, figsize=(15, 10))
plt.suptitle('Histograms of the Columns -Part 2')
plt.show()
```

```python
numerical_columns_subset = df.select_dtypes(include=['int64', 'float64']).columns[31:]  

# Create histograms for the subset of numerical columns
df[numerical_columns_subset].hist(bins=20, figsize=(15, 10))
plt.suptitle('Histograms of the Columns -Part 3')
plt.show()
```

```python
correlation_matrix = df.corr()
```

```python
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Map of Numerical Variables')
plt.show()
```
